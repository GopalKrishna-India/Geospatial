**To** **download data for multiple dates**

```python
# Define the dates you want to retrieve records for
target_dates = ["11_1_2024", "12_2_2024", "13_2_2024"]

# Reference to the root of the database
root_ref = db.child("/UsersAfter07Jul23")

# Retrieve all records from the database
all_records = root_ref.get().val()

# Filter records for the specified dates
records_for_dates = {}

for target_date in target_dates:
    records_for_dates[target_date] = {}
    for key, value in all_records.items():
        if target_date in key:
            records_for_dates[key] = value

print(records_for_dates)
```
