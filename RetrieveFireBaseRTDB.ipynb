**To Download Data of a given Date**

```python
# Define the date you want to retrieve records for
target_date = "13_9_2023"#, 9_9_2023, 10_9_2023, 11_9_2023, 12_9_2023, 13_9_2023, 14_9_2023, 15_9_2023

# Reference to the root of the database
root_ref = db.child("/UsersAfter07Jul23")

# Retrieve all records from the database
all_records = root_ref.get().val()

## Filter records for the specified date
records_for_date = {}

for key, value in all_records.items():
    if target_date in key:
        records_for_date[key] = value

print(records_for_date)
```

**Convert dataframe to JSON**

```python
import json

# Save the filtered records as a JSON file
output_filename = f"AllrecordsTill_{14_9_2023}.json"
#output_filename = "C:/Users/Gopal Krishna/Geotagging Data/14_Aug23_Records.json"

with open(output_filename, "w") as json_file:
    json.dump(records_for_date, json_file, indent=4)

print(f"Filtered records saved to {output_filename}")
```
